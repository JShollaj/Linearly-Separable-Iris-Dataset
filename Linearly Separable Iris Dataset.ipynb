```python
# load libraries
from sklearn.svm import LinearSVC
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Load data with only two classes and two features
# so that we can visualize them later
iris = datasets.load_iris()
features = iris.data[:100, :2]
target = iris.target[:100]
```

```python
# display features and target (flower type)
print(features.shape, target.shape)
print(features[0:4])
print(target[:])
```

```python
# Standardize features
scaler = StandardScaler()
features_standardized = scaler.fit_transform(features)
print(features_standardized[0:4])
```

```python
# Create support vector classifier
svc = LinearSVC(C=1.0)
```

```python
# Train model
model = svc.fit(features_standardized, target)
```

```python
# Plot data points and color using their class
color = ["red" if c == 0 else "lightgrey" for c in target]
plt.scatter(features_standardized[:,0], features_standardized[:,1], c=color)
```

```python
# Create the hyperplane
w = svc.coef_[0]
a = -w[0] / w[1]
xx = np.linspace(-2.5, 2.5)
yy = a * xx - (svc.intercept_[0]) / w[1]
```

```python
# Plot data points and color using their class
color = ["red" if c == 0 else "lightgrey" for c in target]
plt.scatter(features_standardized[:,0], features_standardized[:,1], c=color)
# Plot the hyperplane
plt.plot(xx, yy)
plt.axis("off"), plt.show();
```

```python
# Create two new observations: top-left and bottom-right corner of visualization
new_observations = [[-2, 3], [2, -3]]
```

```python
# Predict class of new observations
svc.predict(new_observations)
```
